# Evaluation
Generate plots, accuracies, etc. The good stuff.

The actual evaluation is on the Excel files because it is way easier than in Python.
This code reads the predictions (that you should have already reshaped with `reshape_predictions.py`), and dumps an Excel file with shot-by-delta averages.

The Excel file also contains averages of these averages plus error analysis on call failures. Check the ones from the repo to see how the rest of the analysis was carried out. 

```python
from eval_utils import inject_stats_sheets, print_table

problem2fname = {"PARITY": "parity",
                 "Pattern_Matching": "pattern_matching",
                 "Reversal": "reversal",
                 "Stack": "stack",
                 "Vending_Machine": "vending_machine",
                 "Vending_Machine_Sum": "vending_machine_with_sum",
                 "MazeComplete": "maze_complete",
                 "MazeSolve": "maze_solve",
                 "Hamiltonian": "hamiltonian",
                }

fname2problem = {v: k for k, v in problem2fname.items()}

model2printname = {"turbo": "GPT-4 Turbo",
                    "omni": "GPT-4o",
                    "o3-mini": "o3-mini",
                    "qwen": "Qwen-2B",
                    "phi3": " Phi-3.5 MoE",
                    "mixtral": "Mixtral 7x8B",
                    "omni-imbalanced-labels-consolidated": "GPT-4o (imbalanced labels)", 
                    "omni-random-exemplars": "GPT-4o (fully random exemplars)",
                    "omni-random-labels-consolidated": "GPT-4o (random labels)"}

mode2printname = {"modus_ponens": "Modus ponens",
                  "w_desc": "With description",
                  "apo": "APO",
                  "cot": "CoT",
                  "sot": "SoT",
                  "w_automaton": "Automaton",
                  "word_salad": "Word salad"}
```

```python
models = ["omni-imbalanced-labels-consolidated", "omni-random-exemplars", "omni-random-labels-consolidated", "omni"]
modes = ["modus_ponens", "w_desc", "w_automaton", "word_salad", "apo", "cot", "sot"]
filename = "all_results_ablation_unpunished_new.xlsx"
rename_gpt4o = True
punish = False

for k, v in problem2fname.items():
    prompts_acc, prompts_err = {}, {}
    row = 0
    for mode in modes: 
        row, data = print_table(k, mode, models, problem2fname, model2printname, 
                                shot_avg=True, do_print_table=True, rename_gpt4o=rename_gpt4o, 
                                punish=punish, excel_filename=filename, row=row)
print_models = [model2printname[m] for m in models]
if rename_gpt4o:
    print_models[print_models.index("GPT-4o")] = "GPT-4o (random exemplars)"

if filename is not None:
    inject_stats_sheets([k for k in problem2fname.keys()], modes, print_models, filename)
```

## Stats

This loads the datasets and prints stats for every single corpus and outputs the statistics from the appendix.
The beautiful thing (which I should have done earlier) is that this one prints it in LaTeX.

```python
import json
from glob import glob
import numpy as np

# This is a quick and dirty way to semi-print a LateX table with stats.
dirs = ["PARITY", "Pattern_Matching", "Reversal", "Stack", 
        "Hamiltonian", "MazeComplete", "MazeSolve", "Vending_Machine"]

header_str = """
        \\begin{tabular}{l|l|l}
        \\textbf{Task} & \\textbf{Average Lengths} ($\\delta$ = 0, 0.2, 0.45, 0.65, 0.85) & Label Balance (\%) \\\\\\midrule"""

print(header_str)

for dir in dirs:
    test_sets = [f for f in glob(f"{dir}/*test*.json") if "shuffled_labels" not in f]
    lengths = {}
    additional_stats = {}
    for t in test_sets:
        if dir == "PARITY" and "ood" in t and "negs" not in t: continue
        lines = [json.loads(l) for l in open(t, "r", encoding="utf-8").readlines()]
        data_dict = {"delta_0": lines} if "id" in t else lines[0]
        for key, dataset in data_dict.items():
            lengths[key] = {
                "Lengths": [len(l["Entry"]) for l in dataset] if "Hamiltonian" in dir else [len(l["Entry"]) for l in dataset],
                "Labels": [l["Label"] for l in dataset]
            }
            if "Hamiltonian" in dir:
                additional_stats[key] = {
                    "Vertices": [l["Vertices"] for l in dataset],
                    "Path": [len(l["Path"]) for l in dataset]
                }

    deltas = ["delta_0", 'delta_0.2', 'delta_0.45', 'delta_0.65', 'delta_0.85']

    get_avg = lambda x, y, z: ", ".join([str(int(np.average(x[delta][y])*z)) for delta in deltas])
    
    label_balance = get_avg(lengths, "Labels", 100)
    len_str = get_avg(lengths, "Lengths", 1)
    if "Hamiltonian" not in dir:
        print(f"{dir} & {len_str} & {label_balance} \\\\")
    else:
        ver_str = get_avg(additional_stats, "Vertices", 1)
        pat_str = get_avg(additional_stats, "Path", 1)
        print(f"{dir} & {len_str} (Graph) & {label_balance} \\\\")
        print(f" & {ver_str} (Vertices) & \\\\")
        print(f" & {pat_str} (Paths) & \\\\")

print("\\end{tabular}")
```
